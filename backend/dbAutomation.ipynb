```python
mainCsv['Size'] = mainCsv['Size'].str.replace("M", "")
mainCsv['Size'] = mainCsv['Size'].str.replace("G", "")
mainCsv['Size'] = mainCsv['Size'].str.replace(",", "")
mainCsv['Size'] = mainCsv['Size'].str.replace('Varies with device', "10")
appSize = mainCsv['Size'].str.replace("k", "").astype(float)
print(appSize)
```

```python
mainCsv['Minimum Android'] = mainCsv['Minimum Android'].str.replace(
    " and up", "")
mainCsv['Minimum Android'] = mainCsv['Minimum Android'].str.replace(
    "Varies with device", "10.0.0")
mainCsv['Minimum Android'] = mainCsv['Minimum Android'].str.replace("W", "")
mainCsv['Minimum Android'] = mainCsv['Minimum Android'].str[0:3]
mainCsv['Minimum Android'] = mainCsv['Minimum Android'].fillna("10.0.0")
minAndroid = mainCsv['Minimum Android'].astype(float)

print(minAndroid)
```
